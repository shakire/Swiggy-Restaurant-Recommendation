```python
import pandas as pd
```

```python
df=pd.read_csv("C:\Data science\Project 4 - Swiggy Analysis\swiggy.csv")
```

```python
pd.set_option('display.max_rows', None)       
pd.set_option('display.max_columns', None)        

df.head(10)
```

```python
df.info()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['cost'] = df['cost'].str.replace('₹','')
df['cost'] = pd.to_numeric(df['cost'], errors='coerce')
```

```python
df['cost'] = pd.to_numeric(df['cost'], errors='coerce').astype('Int64')
```

```python
df['rating_count'] = df['rating_count'].str.replace('+','',regex=False)
df['rating_count'] = df['rating_count'].str.replace('ratings','',regex=False)
df['rating_count'] = df['rating_count'].replace('Too Few Ratings',0)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce').astype('Int64')
```

```python
df['city'] = df['city'].str.split(',').str[-1].str.strip()
```

```python
df['city'].unique()
```

```python
df.drop(["id","lic_no","link","menu","address"],axis="columns",inplace=True)
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df = df.drop_duplicates()
```
